```python
# Load the KMeans model and scaler from pickle files
def load_model_and_scaler():
    with open('kmeans_model.pkl', 'rb') as f:
        kmeans_loaded = pickle.load(f)
    with open('scaler.pkl', 'rb') as f:
        scaler_loaded = pickle.load(f)
    return kmeans_loaded, scaler_loaded
```

```python
# Function to retrieve audio features for a track from Spotify
def get_audio_features(track_name, sp):
    """
    Retrieves audio features for a track from Spotify's API.
    """
    # Search for the track by name
    results = sp.search(q=track_name, type='track', limit=5)
    
    if not results['tracks']['items']:
        print("Song not found!")
        return None
    
    # Extract the track ID of the first track
    track_id = results['tracks']['items'][0]['id']
    track_href = results['tracks']['items'][0]['href']
    
    # Get audio features for the track
    features = sp.audio_features(track_id)[0]
    
    # Extracting relevant audio features
    features_data = {
        'track_name': track_name,
        'track_id': track_id,
        'danceability': features['danceability'],
        'energy': features['energy'],
        'key': features['key'],
        'loudness': features['loudness'],
        'mode': features['mode'],
        'speechiness': features['speechiness'],
        'acousticness': features['acousticness'],
        'instrumentalness': features['instrumentalness'],
        'liveness': features['liveness'],
        'valence': features['valence'],
        'tempo': features['tempo'],
        'type': features['type'],
        'uri': features['uri'],
        'track_href': track_href,
        'analysis_url': features['analysis_url'],
        'duration_ms': features['duration_ms'],
        'time_signature': features['time_signature']
    }
    
    return features_data
```

```python
features_data = get_audio_features("What", sp)
features_data
```

```python
# Function to recommend a song from the same cluster
def recommend_song_from_cluster(features_data, sp, spotify_list):
    """
    Recommends a song from the same cluster as the user input song.
    """
    # Load the KMeans model and scaler
    kmeans_loaded, scaler_loaded = load_model_and_scaler()

    # Extract relevant features from song data
    features = [
        features_data['danceability'],
        features_data['energy'],
        features_data['key'],
        features_data['loudness'],
        features_data['mode'],
        features_data['speechiness'],
        features_data['acousticness'],
        features_data['instrumentalness'],
        features_data['liveness'],
        features_data['valence'],
        features_data['tempo']
    ]

    # Convert to a 2D array (1 sample with multiple features)
    song_features = np.array([features])

    # Scale the features using the loaded scaler
    scaled_song_features = scaler_loaded.transform(song_features)

    # Predict the cluster using the loaded KMeans model
    predicted_cluster = kmeans_loaded.predict(scaled_song_features)
    print(f"The song belongs to cluster {predicted_cluster[0]}")

    # Filter rows in spotify_list where the cluster matches the predicted cluster
    cluster_songs = spotify_list[spotify_list['cluster'] == predicted_cluster[0]]

    # Check if there are any matching rows
    if cluster_songs.empty:
        print(f"No songs found in cluster {predicted_cluster[0]}. Unable to recommend a song.")
        return None
    else:
        # Pick a random row from the matching cluster songs
        random_row = cluster_songs.sample(1).iloc[0]
        
        # Retrieve the track_id from the random row
        track_id = random_row['track_id']
        print(f"Recommended track_id from the same cluster: {track_id}")
        return track_id
```

```python
recommend_song_from_cluster(features_data, sp, spotify_list)
```

```python
def play_song(track_id):
    """
    Generates an embeddable Spotify player for the given track ID.
    """
    iframe = IFrame(
        src=f"https://open.spotify.com/embed/track/{track_id}",
        width="320",
        height="80",
        frameborder="0",
        allowtransparency="true",
        allow="encrypted-media"
    )
    display(iframe)
```

```python
from fuzzywuzzy import process

# Function to suggest similar songs and decide if a match is close enough
def suggest_and_play_song(user_input, song_list, sp):
    """
    Suggests songs that closely match the user's input using fuzzy matching.
    If a match is sufficiently close, it plays the song.
    """
    matches = process.extract(user_input, song_list, limit=1)  # Top 1 match
    if matches:
        closest_match, score = matches[0]
        print(f"Did you mean '{closest_match}'? (Score: {score})")

        # Assume the user meant the closest match if the score is high
        if score >= 80:  # Threshold for accepting the suggestion
            print(f"Playing '{closest_match}' as it seems to be a close match.")
            # Search for the song on Spotify
            results = sp.search(q=closest_match, type='track', limit=1)
            if results['tracks']['items']:
                track_id = results['tracks']['items'][0]['id']
                play_song(track_id)  # Play the track
            else:
                print("Could not find the song on Spotify.")
        else:
            print(f"Sorry, I couldn't confidently guess the song. Closest match: '{closest_match}'.")
    else:
        print("No similar songs found.")
```

```python
from fuzzywuzzy import process

# Function to suggest similar songs using fuzzy matching
def suggest_similar_songs(user_input, song_list):
    """
    Suggests songs that closely match the user's input using fuzzy matching.
    """
    matches = process.extract(user_input, song_list, limit=1)  # Top 5 matches
    print("Sorry, I couldn't find an exact match, but here are some similar songs:")
    
    for match in matches:
        print(f"- {match[0]}")

# Main logic for song search and recommendation
if not user_input:
    print("Please enter a song")
else:
    # Convert user_input to lowercase to handle case-insensitive comparison
    user_input_lower = user_input.lower()
    
    # Check if the user input exists in any song title (case insensitive)
    song_match = top_100['Song'].str.lower().str.contains(user_input_lower, na=True)
    
    if song_match.any():
        # Get the artist of the matched song
        artist = top_100.loc[song_match, 'Artist'].values[0]
        
        print(f"This song from {artist} is HOT right now!")
        
        # Find other songs by the same artist, excluding the original song
        artist_songs = top_100[top_100['Artist'] == artist]
        other_songs = artist_songs[artist_songs['Song'].str.lower() != user_input_lower]
        
        if not other_songs.empty:
            # Pick another song by the same artist
            another_song = other_songs.iloc[0]['Song']
            print(f"Another amazing song in the Billboard Hot 100 by {artist} is {another_song}")
        else:
            print(f"No other songs by {artist} in the Billboard Hot 100.")
    else:
        print("This song is not in the Billboard Hot 100 currently.")
        
        # Use fuzzy matching to suggest similar songs
        suggest_similar_songs(user_input, top_100['Song'].tolist())
        
        # Use Spotify API and machine learning-based recommendation flow
        features_data = get_audio_features(user_input, sp)
        
        if features_data:
            # Recommend a song from the same cluster
            recommended_track_id = recommend_song_from_cluster(features_data, sp, spotify_list)
            
            if recommended_track_id:
                print("Here's a recommended song from the same cluster:")
                play_song(recommended_track_id)
            else:
                print("Sorry, no suitable recommendation found.")
        else:
            print("Could not find the song on Spotify.")
```
